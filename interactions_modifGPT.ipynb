```python
import pulp
import holoviews as hv
import panel as pn
import numpy as np
import pandas as pd
hv.extension('bokeh')
```

```python
# Chargement des donnes
df_solar = pd.read_csv('solar.csv')
df_wind = pd.read_csv('wind_onshore.csv')
df_demand = pd.read_csv('demand2050_ADEME.csv', header=None)
df_demand.columns = ["heures", "demande"]

# Calcul consommation annuelle
annual_demand = df_demand['demande'].sum()

# Calcul production potentielle pour 1GW install
annual_solar_per_gw = df_solar['facteur_charge'].sum()
annual_wind_per_gw = df_wind['facteur_charge'].sum()

# Pour avoir 100% de surproduction et un mix 50-50
target_production = annual_demand * 3.5
wind_capacity = target_production/(2 * annual_wind_per_gw)
solar_capacity = target_production/(2 * annual_solar_per_gw)

T = 2160
N = len(df_demand['demande'].values[:3*T])

# Production ENR horaire
prod = {t: wind_capacity * df_wind['facteur_charge'].values[:3*T][t] + solar_capacity * df_solar['facteur_charge'].values[:3*T][t] for t in range(N)}

print(prod)
```

```python
def to_array_from_prod_dict(prod_dict):
    """Convertit le dict {t: value} en numpy array ordonn."""
    N = len(prod_dict)
    return np.array([prod_dict[t] for t in range(N)])

def energy_flows(prodA, prodB, prodC, demandA, demandB, demandC, qmax):
    """
    Rsout un modle linaire :
      prodX[t] - demandX[t] - exports + imports + defX[t] >= 0
    Variables:
      qXY[t] >= 0 flux de X vers Y
      defX[t] >= 0 dficit sur X (ce qu'on veut minimiser)
    Retour: dictionnaire avec statuts et vecteurs de flux/dficits.
    """
    # Assurer que les entres sont numpy arrays et de mme longueur
    prodA = np.asarray(prodA)
    prodB = np.asarray(prodB)
    prodC = np.asarray(prodC)
    demandA = np.asarray(demandA)
    demandB = np.asarray(demandB)
    demandC = np.asarray(demandC)
    assert len(prodA) == len(demandA)
    assert len(prodB) == len(demandB)
    assert len(prodC) == len(demandC)
    N = len(prodA)

    # Problme
    prob = pulp.LpProblem("Energy_flows", pulp.LpMinimize)

    # Variables de flux (continues, >=0)
    qAB = pulp.LpVariable.dicts("qAB", range(N), upBound=qmax)
    qAC = pulp.LpVariable.dicts("qAC", range(N), upBound=qmax)
    qBA = pulp.LpVariable.dicts("qBA", range(N), upBound=qmax)
    qBC = pulp.LpVariable.dicts("qBC", range(N), upBound=qmax)
    qCA = pulp.LpVariable.dicts("qCA", range(N), upBound=qmax)
    qCB = pulp.LpVariable.dicts("qCB", range(N), upBound=qmax)

    # Variables de dficit (>=0)  ce que l'on minimise
    defA = pulp.LpVariable.dicts("defA", range(N), lowBound=0)
    defB = pulp.LpVariable.dicts("defB", range(N), lowBound=0)
    defC = pulp.LpVariable.dicts("defC", range(N), lowBound=0)

    # Objectif : minimiser la somme des dficits horaires (option : pondrer si besoin)
    prob += pulp.lpSum([defA[t] + defB[t] + defC[t] for t in range(N)])

    # Contraintes horaires (bilan) :
    # prodA - demandA - qAB - qAC + qBA + qCA + defA >= 0
    # -> defA >= demandA - (prodA - qAB - qAC + qBA + qCA)
    for t in range(N):
        prob += (prodA[t] - demandA[t]
                 - qAB[t] - qAC[t]
                 + qBA[t] + qCA[t]
                 + defA[t] >= 0)
        prob += (prodB[t] - demandB[t]
                 - qBA[t] - qBC[t]
                 + qAB[t] + qCB[t]
                 + defB[t] >= 0)
        prob += (prodC[t] - demandC[t]
                 - qCA[t] - qCB[t]
                 + qAC[t] + qBC[t]
                 + defC[t] >= 0)

    # Contraintes de capacit des interconnexions (par heure)
    for t in range(N):
        prob += (prodA[t] - demandA[t])*qAB[t] <= 0
        prob += (prodA[t] - demandA[t])*qAC[t] <= 0
        prob += (prodB[t] - demandB[t])*qBA[t] <= 0
        prob += (prodB[t] - demandB[t])*qBC[t] <= 0
        prob += (prodC[t] - demandC[t])*qCA[t] <= 0
        prob += (prodC[t] - demandC[t])*qCB[t] <= 0

    # Rsoudre
    prob.solve()  # msg=True pour logs

    # Rcuprer rsultats (convertir en numpy arrays)
    def get_vec(var_dict):
        return np.array([var_dict[t].value() for t in range(N)], dtype=float)

    result = {
        'status': pulp.LpStatus[prob.status],
        'qAB': get_vec(qAB),
        'qAC': get_vec(qAC),
        'qBA': get_vec(qBA),
        'qBC': get_vec(qBC),
        'qCA': get_vec(qCA),
        'qCB': get_vec(qCB),
        'defA': get_vec(defA),
        'defB': get_vec(defB),
        'defC': get_vec(defC),
        'total_deficit': sum(get_vec(defA)) + sum(get_vec(defB)) + sum(get_vec(defC)),
    }
    return result
```

```python
prod_arr = to_array_from_prod_dict(prod)

# Lancer l'optimisation
result = energy_flows(
    prodA = prod_arr[0:T], 
    prodB = prod_arr[T:2*T], 
    prodC = prod_arr[2*T : 3*T], 
    demandA = df_demand['demande'].values[0:T], 
    demandB = df_demand['demande'].values[T:2*T], 
    demandC = df_demand['demande'].values[2*T:3*T], 
    qmax = 17e12
)



# Analyse des rsultats
print("Status:", result['status'])
print("Total deficit (sum over all hours and zones):", result['total_deficit'])
print("Sum flows qAB (GWh):", result['qAB'].sum())
print("Quelques valeurs qAB[:10]:", result['qAB'][:10])
print("defA[:10]:", result['defA'][:10])

```
