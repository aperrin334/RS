```python
import pulp
import holoviews as hv
import panel as pn
import numpy as np
import pandas as pd
hv.extension('bokeh')
```

```python
# Chargement des données
df_solar = pd.read_csv('solar.csv')
df_wind = pd.read_csv('wind_onshore.csv')
df_demand = pd.read_csv('demand2050_ADEME.csv', header=None)
df_demand.columns = ["heures", "demande"]

# Calcul consommation annuelle
annual_demand = df_demand['demande'].sum()

# Calcul production potentielle pour 1GW installé
annual_solar_per_gw = df_solar['facteur_charge'].sum()
annual_wind_per_gw = df_wind['facteur_charge'].sum()

# Pour avoir 100% de surproduction et un mix 50-50
target_production = annual_demand * 3.5
wind_capacity = target_production/(2 * annual_wind_per_gw)
solar_capacity = target_production/(2 * annual_solar_per_gw)

print(f"Capacités requises: éolien = {wind_capacity:.1f}GW, solaire = {solar_capacity:.1f}GW")
```

```python
def optimize_storage(solar_profile, wind_profile, demand, wind_cap, solar_cap,
                    phs_capacity=180, phs_power=9.3, phs_efficiency=0.75,
                    ptg_capacity=125000, ptg_power_in=7.66, ptg_power_out=32.93, 
                    ptg_efficiency=0.4):
    """
    Optimise la stratégie de stockage sur l'année avec contrainte cyclique
    
    Parameters:
    -----------
    solar_profile, wind_profile: séries temporelles des facteurs de charge
    demand: série temporelle de la demande
    wind_cap, solar_cap: capacités installées en GW
    phs_capacity: capacité de stockage PHS en GWh
    phs_power: puissance max PHS en GW
    phs_efficiency: rendement PHS (aller-retour)
    ptg_capacity: capacité de stockage P2G en GWh
    ptg_power_in: puissance max électrolyse en GW  
    ptg_power_out: puissance max reconversion en GW
    ptg_efficiency: rendement P2G (aller-retour)
    """
    
    N = len(demand)  # nombre d'heures
    
    # Création du problème
    prob = pulp.LpProblem("Storage_Strategy", pulp.LpMinimize)
    
    # Variables
    phs_in = pulp.LpVariable.dicts("phs_in", range(N), 0)  # Charge PHS
    phs_out = pulp.LpVariable.dicts("phs_out", range(N), 0)  # Décharge PHS
    phs_level = pulp.LpVariable.dicts("phs_level", range(N), 0)  # Niveau PHS
    
    ptg_in = pulp.LpVariable.dicts("ptg_in", range(N), 0)  # Charge P2G
    ptg_out = pulp.LpVariable.dicts("ptg_out", range(N), 0)  # Décharge P2G
    ptg_level = pulp.LpVariable.dicts("ptg_level", range(N), 0)  # Niveau P2G
    
    deficit = pulp.LpVariable.dicts("deficit", range(N), 0)  # Déficit
    
    # Production ENR horaire
    prod = {t: wind_cap * wind_profile[t] + solar_cap * solar_profile[t] 
            for t in range(N)}
    
    # Fonction objectif : minimiser déficits
    prob += -ptg_level[N-1] + 1e4 * pulp.lpSum(deficit[t] for t in range(N))
    
    # Contraintes
    
    # Niveaux initiaux
    prob += phs_level[0] == phs_capacity/2  # PHS à 50%
    prob += ptg_level[0] == ptg_capacity    # P2G plein
    
    # Contraintes cycliques : les niveaux finaux doivent être égaux aux niveaux initiaux
    #prob += phs_level[N-1] == phs_level[0]
    #prob += ptg_level[N-1] == ptg_level[0]
    
    for t in range(N):
        # Équilibre offre-demande
        prob += prod[t] + phs_out[t]*phs_efficiency + ptg_out[t]*ptg_efficiency - \
                phs_in[t] - ptg_in[t] + deficit[t] >= demand[t]
        
        # Limites PHS
        prob += phs_in[t] <= phs_power
        prob += phs_out[t] <= phs_power
        prob += phs_level[t] <= phs_capacity
        
        # Limites P2G
        prob += ptg_in[t] <= ptg_power_in
        prob += ptg_out[t] <= ptg_power_out
        prob += ptg_level[t] <= ptg_capacity
        
        # Évolution des stocks
        if t > 0:
            prob += phs_level[t] == phs_level[t-1] + phs_in[t]*phs_efficiency - phs_out[t]
            prob += ptg_level[t] == ptg_level[t-1] + ptg_in[t]*ptg_efficiency - ptg_out[t]
       
        
    # Résolution
    prob.solve()
    
    return {
        'status': pulp.LpStatus[prob.status],
        'phs_in': [phs_in[t].value() for t in range(N)],
        'phs_out': [phs_out[t].value() for t in range(N)],
        'phs_level': [phs_level[t].value() for t in range(N)],
        'ptg_in': [ptg_in[t].value() for t in range(N)],
        'ptg_out': [ptg_out[t].value() for t in range(N)],
        'ptg_level': [ptg_level[t].value() for t in range(N)],
        'deficit': [deficit[t].value() for t in range(N)]
    }
```

```python
import numpy as np
from scipy.optimize import linprog
from scipy.sparse import vstack, csr_matrix

def optimize_storage_linprog(solar_profile, wind_profile, demand, wind_cap, solar_cap,
                                    phs_capacity=180, phs_power=9.3, phs_efficiency=0.75,
                                    ptg_capacity=125000, ptg_power_in=7.66, ptg_power_out=32.93,
                                    ptg_efficiency=0.4):
    """
    Version corrigée ET optimisée mémoire :
    - utilise des matrices creuses (sparse)
    - reproduit fidèlement la formulation du modèle PuLP
    """

    N = len(demand)
    M = 7 * N  # nombre total de variables

    # Indices
    idx_phs_in    = 0
    idx_phs_out   = N
    idx_phs_level = 2 * N
    idx_ptg_in    = 3 * N
    idx_ptg_out   = 4 * N
    idx_ptg_level = 5 * N
    idx_deficit   = 6 * N

    # --- Fonction objectif ---
    c = np.zeros(M)
    c[idx_deficit:idx_deficit + N] = 1e4             # pénalisation déficit
    c[idx_ptg_level + (N - 1)] = -1.0                # maximiser ptg_level final

    # --- Contraintes ---
    A_ub_rows = []
    b_ub = []
    A_eq_rows = []
    b_eq = []

    # Production ENR
    prod = np.array([wind_cap * wind_profile[t] + solar_cap * solar_profile[t] for t in range(N)])

    # Équilibre offre-demande
    for t in range(N):
        row = np.zeros(M)
        row[idx_phs_in + t] = 1.0
        row[idx_phs_out + t] = -phs_efficiency
        row[idx_ptg_in + t] = 1.0
        row[idx_ptg_out + t] = -ptg_efficiency
        row[idx_deficit + t] = -1.0
        A_ub_rows.append(csr_matrix(row))
        b_ub.append(prod[t] - demand[t])

    # Limites de puissance PHS
    for t in range(N):
        row = np.zeros(M)
        row[idx_phs_in + t] = 1.0
        A_ub_rows.append(csr_matrix(row)); b_ub.append(phs_power)
        row = np.zeros(M)
        row[idx_phs_out + t] = 1.0
        A_ub_rows.append(csr_matrix(row)); b_ub.append(phs_power)

    # Limites de puissance P2G
    for t in range(N):
        row = np.zeros(M)
        row[idx_ptg_in + t] = 1.0
        A_ub_rows.append(csr_matrix(row)); b_ub.append(ptg_power_in)
        row = np.zeros(M)
        row[idx_ptg_out + t] = 1.0
        A_ub_rows.append(csr_matrix(row)); b_ub.append(ptg_power_out)

    # Contraintes de capacité sur les niveaux
    for t in range(N):
        row = np.zeros(M)
        row[idx_phs_level + t] = 1.0
        A_ub_rows.append(csr_matrix(row)); b_ub.append(phs_capacity)
        row = np.zeros(M)
        row[idx_ptg_level + t] = 1.0
        A_ub_rows.append(csr_matrix(row)); b_ub.append(ptg_capacity)

    # Conditions initiales
    row = np.zeros(M); row[idx_phs_level + 0] = 1.0
    A_eq_rows.append(csr_matrix(row)); b_eq.append(phs_capacity / 2.0)

    row = np.zeros(M); row[idx_ptg_level + 0] = 1.0
    A_eq_rows.append(csr_matrix(row)); b_eq.append(ptg_capacity)

    # Évolution des niveaux (t > 0)
    for t in range(1, N):
        # PHS
        row = np.zeros(M)
        row[idx_phs_level + t]     = 1.0
        row[idx_phs_level + t - 1] = -1.0
        row[idx_phs_in + t]        = -phs_efficiency
        row[idx_phs_out + t]       = 1.0
        A_eq_rows.append(csr_matrix(row)); b_eq.append(0.0)

        # P2G
        row = np.zeros(M)
        row[idx_ptg_level + t]     = 1.0
        row[idx_ptg_level + t - 1] = -1.0
        row[idx_ptg_in + t]        = -ptg_efficiency
        row[idx_ptg_out + t]       = 1.0
        A_eq_rows.append(csr_matrix(row)); b_eq.append(0.0)

    # Contraintes cycliques :
    # row = np.zeros(M); row[idx_phs_level + (N - 1)] = 1.0; row[idx_phs_level + 0] = -1.0
    # A_eq_rows.append(csr_matrix(row)); b_eq.append(0.0)
    # row = np.zeros(M); row[idx_ptg_level + (N - 1)] = 1.0; row[idx_ptg_level + 0] = -1.0
    # A_eq_rows.append(csr_matrix(row)); b_eq.append(0.0)

    # Conversion en matrices creuses
    A_ub = vstack(A_ub_rows) if A_ub_rows else None
    b_ub = np.array(b_ub) if b_ub else None
    A_eq = vstack(A_eq_rows) if A_eq_rows else None
    b_eq = np.array(b_eq) if b_eq else None

    # Bornes (toutes >= 0)
    lb = np.zeros(M)
    ub = np.full(M, np.inf)
    bounds = list(zip(lb, ub))

    # Résolution
    res = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')

    if not res.success:
        raise ValueError("Optimization failed: " + res.message)

    x = res.x
    phs_in    = x[idx_phs_in : idx_phs_in + N]
    phs_out   = x[idx_phs_out : idx_phs_out + N]
    phs_level = x[idx_phs_level : idx_phs_level + N]
    ptg_in    = x[idx_ptg_in : idx_ptg_in + N]
    ptg_out   = x[idx_ptg_out : idx_ptg_out + N]
    ptg_level = x[idx_ptg_level : idx_ptg_level + N]
    deficit   = x[idx_deficit : idx_deficit + N]

    return {
        'status': res.message,
        'phs_in': phs_in,
        'phs_out': phs_out,
        'phs_level': phs_level,
        'ptg_in': ptg_in,
        'ptg_out': ptg_out,
        'ptg_level': ptg_level,
        'deficit': deficit,
        'sol': res
    }
```

```python
# Lancer l'optimisation
T = 8760
result = optimize_storage(
    df_solar['facteur_charge'].values[:T],
    df_wind['facteur_charge'].values[:T],
    df_demand['demande'].values[:T],
    wind_cap=wind_capacity,
    solar_cap=solar_capacity
)

# Analyse des résultats
print(f"Status: {result['status']}")
print(f"Déficit total: {sum(result['deficit']):.1f} GWh")
print(f"PHS final: {result['phs_level'][-1]:.1f} GWh")
print(f"P2G final: {result['ptg_level'][-1]:.1f} GWh")

```
